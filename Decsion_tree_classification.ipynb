```python
train_df['Sex'] = train_df['Sex'].apply(lambda x : 0 if x=='male' else 1)
```

```python
y= train_df['Survived'].values
```

```python
y
```

```python
columns=['Pclass','Sex','Age','SibSp']
feature = train_df[list(columns)].values
```

```python
feature
```

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
X = imp.fit_transform(feature)
```

```python
X
```

```python
from sklearn import tree
my_tree_one = tree.DecisionTreeClassifier(criterion='entropy',max_depth=3, random_state=1)
```

```python
my_tree_one = my_tree_one.fit(X,y)
```

```python
my_tree_one.score(X,y)
```
